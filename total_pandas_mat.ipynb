```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([2023,2024,2025,2026])
y1 = np.array([10,15,30,20])
y2 = np.array([10,23,40,223])
y3 = np.array([12,13,14,18])
line_style = dict( marker="o",markersize=10,markerfacecolor="cyan",markeredgecolor="orange",linestyle="solid",linewidth=2)
plt.plot(x,y1,color="green",**line_style)
plt.plot(x,y2,color="yellow",**line_style)
plt.plot(x,y3,color="brown",**line_style)
plt.show()
```

```python
#Barchart
categories = np.array(["Grains","Fruit","Vegetables","protein","Dairy","sweets"])
values = np.array([4,3,2,5,3,1])
plt.barh(categories,values,color="skyblue")
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()
```

```python
categories = ["Freshmen","sophomores","juniors","seniors"]
values = np.array([300,250,275,225])
colors = ["red","green","blue","orange"]
plt.title("piechart",fontsize=30)
plt.pie(values,labels=categories,autopct="%1.1f%%",explode=[0,0,0,0],shadow=True,startangle=189)
plt.show()
```

```python
x1 = [0,1,1,2,3,4,5,6,7,7,8]
y1 = [55,60,65,62,68,70,75,78,82,85,87]
x2 = [8,7,7,6,5,4,3,2,1,1,0]
y2 = [87,85,82,78,75,70,68,62,65,60,55]
plt.scatter(x1,y1,color="yellow",alpha=0.5,label="class A")
plt.scatter(x1,y1,color="red",alpha=0.5,label="class B")

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")
plt.legend()
plt.show()
```
